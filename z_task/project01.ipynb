```python
import pandas as pd

# Ï£ºÎ≥Ñ ÌèâÍ∑† Ïó∞Í∞Ñ ÏÜåÎìù Í≥ÑÏÇ∞
lc_an_inc_df = lc_df.groupby('addr_state')['annual_inc'].mean().reset_index()

# Í∞í ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨
lc_an_inc_df = lc_an_inc_df.sort_values(by='annual_inc', ascending=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ÎåÄÏ∂ú ÏÉÅÌÉúÏóê Îî∞Î•∏ ÏÉÅÌôòÍ≥º ÌöåÏàòÏú® Í≥ÑÏÇ∞ Ìï®Ïàò Ï†ïÏùò
def calculate_repayment_and_recovery_rate(df):
    df['repayment_rate'] = df['total_rec_prncp'] / df['total_pymnt']
    df['recovery_rate'] = df['recoveries'] / df['funded_amnt']
    return df

# ÎåÄÏ∂ú Îç∞Ïù¥ÌÑ∞ÏÖãÏóêÏÑú ÌïÑÏöîÌïú Ïó¥Îßå ÏÑ†ÌÉùÌïòÏó¨ ÏÉàÎ°úÏö¥ DataFrame ÏÉùÏÑ±
selected_columns = ['total_rec_prncp', 'recoveries', 'funded_amnt', 'total_pymnt','income_level','grade']
selected_df = lc_df[selected_columns].copy()

# ÎåÄÏ∂ú ÏÉÅÌÉúÎ•º Í∏∞Ï§ÄÏúºÎ°ú ÏÉÅÌôòÍ≥º ÌöåÏàòÏú® Í≥ÑÏÇ∞
selected_df = calculate_repayment_and_recovery_rate(selected_df)
# ÏÜåÎìù Íµ¨Í∞ÑÍ≥º Ïã†Ïö© Îì±Í∏âÎ≥ÑÎ°ú ÏÉÅÌôòÍ≥º ÌöåÏàòÏú® Í≥ÑÏÇ∞
repayment_and_recovery_rate_by_grade = selected_df.groupby(['income_level', 'grade']).mean()
```
