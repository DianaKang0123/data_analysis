#### 🎀 각 주별 연간 소득 비교

```python
import pandas as pd

# 주별 평균 연간 소득 계산
lc_an_inc_df = lc_df.groupby('addr_state')['annual_inc'].mean().reset_index()

# 값 내림차순 정렬
lc_an_inc_df = lc_an_inc_df.sort_values(by='annual_inc', ascending=False)
```

```python
# 주별 중위 연간 소득 계산
lc_median_inc_df = lc_df.groupby('addr_state')['annual_inc'].median().reset_index()

# 값 내림차순 정렬
lc_median_inc_df = lc_median_inc_df.sort_values(by='annual_inc', ascending=False)

# 주별 최대 연간 소득 계산
lc_max_inc_df = lc_df.groupby('addr_state')['annual_inc'].max().reset_index()

# 값 내림차순 정렬
lc_max_inc_df = lc_max_inc_df.sort_values(by='annual_inc', ascending=False)

# 주별 최소 연간 소득 계산
lc_min_inc_df = lc_df.groupby('addr_state')['annual_inc'].min().reset_index()

# 값 내림차순 정렬
lc_min_inc_df = lc_min_inc_df.sort_values(by='annual_inc', ascending=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 그래프 생성
fig, axs = plt.subplots(2, 2, figsize=(18, 10))

# 그래프에 사용할 소득 데이터프레임 리스트 생성
income_dataframes = [lc_an_inc_df, lc_median_inc_df, lc_max_inc_df, lc_min_inc_df]

# 그래프 유형 리스트
graph_types = ['Average', 'Median', 'Max', 'Min']

# 색상 설정
colors = ['r', 'g', 'b', 'y']

# 반복문을 통해 각 그래프 그리기
for i, (income_df, graph_type, color) in enumerate(zip(income_dataframes, graph_types, colors)):
    sns.barplot(data=income_df, x='addr_state', y='annual_inc', ax=axs[i // 2, i % 2], color=color)
    axs[i // 2, i % 2].set_xlabel('State')
    axs[i // 2, i % 2].set_ylabel(f'{graph_type} Annual Income')
    axs[i // 2, i % 2].set_title(f'{graph_type} Annual Income by State')
    axs[i // 2, i % 2].tick_params(axis='x', rotation=90)
    
    # 전체 주의 평균 소득 가로선 추가
    total_mean_income = lc_df['annual_inc'].mean()
    axs[i // 2, i % 2].axhline(y=total_mean_income, color='black', linestyle='--', label='Total Mean Income')

# 그래프 간 간격 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

## 소득분위 별 대출 상태

- Current: 현재 상환 중인 대출 
- Fully Paid: 전액 상환된 대출
- Charged Off: 체납된 대출 
- Late (31-120 days): 31일부터 120일까지 연체된 대출
- In Grace Period: 관리 기간 내에 있는 대출 
- Late (16-30 days): 16일부터 30일까지 연체된 대출 
- Default: 기본 상태인 대출

**In Grace Period 와 Default 제외 체납, 연체, 완납, 납부진행중만 비율 비교**

```python
# 'income_level' 열을 기준으로 그룹화하여 대출 상태를 요약
loan_status_by_income = lc_df.groupby('income_level')['loan_status'].value_counts(normalize=True).unstack()
loan_status_by_income['Late'] = loan_status_by_income['Late (16-30 days)'] + loan_status_by_income['Late (31-120 days)']

# 시각화
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(20, 6))

# Current 상태의 max 값을 구함
current_max = loan_status_by_income['Current'].max() + 0.05

for i, (status, color) in enumerate([('Charged Off', 'red'), ('Fully Paid', 'blue'), ('Current', 'orange'), ('Late', 'green')]):
    sns.barplot(data=loan_status_by_income, x=loan_status_by_income.index, y=status, color=color, label=status, ax=axes[i])
    axes[i].set_title(f'{status} by Income Level')
    axes[i].set_xlabel('Income Level')
    axes[i].set_ylabel('Loan Status Ratio')
    axes[i].legend()
    axes[i].tick_params(axis='x', rotation=45)
    
    # y 축 범위를 Current 상태의 max 값으로 설정
    axes[i].set_ylim(0, current_max)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 대출 상태에 따른 상환과 회수율 계산 함수 정의
def calculate_repayment_and_recovery_rate(df):
    df['repayment_rate'] = df['total_rec_prncp'] / df['total_pymnt']
    df['recovery_rate'] = df['recoveries'] / df['funded_amnt']
    return df

# 대출 데이터셋에서 필요한 열만 선택하여 새로운 DataFrame 생성
selected_columns = ['total_rec_prncp', 'recoveries', 'funded_amnt', 'total_pymnt','income_level','grade']
selected_df = lc_df[selected_columns].copy()

# 대출 상태를 기준으로 상환과 회수율 계산
selected_df = calculate_repayment_and_recovery_rate(selected_df)
# 소득 구간과 신용 등급별로 상환과 회수율 계산
repayment_and_recovery_rate_by_grade = selected_df.groupby(['income_level', 'grade']).mean()
```

```python
from IPython.display import display
import folium

# 미국 중심에 지도 생성
map_usa = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# GeoJSON 파일 다운로드
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/us-states.json'
geo_json_data = requests.get(url).json()

# 히트맵 생성
def create_choropleth_map(data, column, legend_name):
    choropleth_map = folium.Map(location=[37.0902, -95.7129], zoom_start=4)
    folium.Choropleth(
        geo_data=geo_json_data,
        name='choropleth',
        data=data,
        columns=['addr_state', column],
        key_on='feature.id',
        fill_color='RdPu',
        fill_opacity=0.7,
        line_opacity=0.7,
        legend_name=legend_name
    ).add_to(choropleth_map)
    folium.LayerControl().add_to(choropleth_map)
    return choropleth_map

# 주택 소유 형태를 각 주소 주(state)별로 빈도 계산 (ANY 제외)
home_ownership_by_state = lc_df[lc_df['home_ownership'] != 'ANY'].groupby('addr_state')['home_ownership'].value_counts(normalize=True).unstack().reset_index()

# 각각의 지도 생성
map_mortgage = create_choropleth_map(home_ownership_by_state, 'MORTGAGE', 'Mortgage Ratio by State')
map_own = create_choropleth_map(home_ownership_by_state, 'OWN', 'Own Ratio by State')
map_rent = create_choropleth_map(home_ownership_by_state, 'RENT', 'Rent Ratio by State')

# 지도들을 가로로 나란히 표시
print('MORTGAGE')
display(map_mortgage)
print('OWN')
display(map_own)
print('RENT')
display(map_rent)
```
