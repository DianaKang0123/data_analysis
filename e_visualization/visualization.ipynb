```python
import pandas as pd

t_df = pd.read_csv('./datasets/titanic.csv')
t_df.head()
```

```python
t_df.isna().sum()
```

```python
t_df['Pclass'].value_counts()
```

```python
t_df['Age'].describe()
```

```python
t_df.Age = t_df.Age.fillna(t_df.Age.mean())
```

```python
t_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# height 는 max 값이다.
plt.bar(x = t_df.Pclass, height = t_df.Age)
plt.show()

sns.barplot(x='Pclass',y='Age',data=t_df, errorbar=('ci', 40))
plt.show()
```

```python
plt.violinplot(t_df.Age)
plt.show()

# hue로 설정할 feature을 x축으로 설정하면 여러개의 군집을 한번에 볼 수 있다.
sns.violinplot(x='Pclass', y='Age',data=t_df, hue='Pclass')
plt.show()
```

```python
# 산점도
plt.scatter(x=t_df.Age, y=t_df.Fare)
plt.show()

sns.scatterplot(x='Age',y='Fare',hue='Pclass',data=t_df)

t_df.plot(x='Age',y='Fare',kind='scatter')
```

```python
# 히스토그램
# t축은 ccount
# t축은 ccount
sns.histplot(x='Age',data=t_df, hue='Sex', multiple='dodge')
plt.show()

sns.histplot(x='Age',data=t_df, hue='Sex', multiple='stack')
plt.show()
```

```python
sns.histplot(x='Age',data=t_df, kde=True, multiple='stack')
plt.show()
```

```python
plt.hist(t_df.Age)
plt.show()

sns.histplot(t_df.Age, kde=True)
plt.show()

sns.distplot(t_df.Age, kde=True, rug=True)
plt.show()

t_df.Age.hist(grid=False)
```

```python
import pandas as pd

a_df = pd.read_csv('./datasets/avocado.csv')
```

```python
a_df
```

```python
a_df.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.lineplot(x='Date', y='AveragePrice',data=a_df)
```

```python
a_df.loc[:,'Date'] = pd.to_datetime(a_df.Date)
line = sns.lineplot(x='Date', y='AveragePrice',data=a_df)
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()
```

```python
line = sns.lineplot(x='Date', y='AveragePrice',hue ='type', data=a_df)
line.set_xticklabels(line.get_xticklabels(), rotation=90)
plt.show()
```
